Import Libraries

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

Dataset From Kaggle

```python
house = pd.read_csv('house_price_regression_dataset.csv')
house
```

# Polynomial Regression

```python
X = house.iloc[:, :-1].values
y = house.iloc[:, -1].values
```

```python
print(X), print(y)
```

Splitting dataset into Training set and Test set

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state = 0)
```

```python
print(X_train)
print(y_train)
```

```python
print(X_test)
print(y_test)
```

Training the Polynomial Regression model on the whole dataset

```python
from sklearn.preprocessing import PolynomialFeatures
from sklearn.linear_model import LinearRegression

poly = PolynomialFeatures(degree=2)
X_poly = poly.fit_transform(X_train)
lin_reg = LinearRegression()
lin_reg.fit(X_poly, y_train)
```

Predicting Results

Predicting a new result with Polynomial Regression

```python
print(lin_reg.predict(poly.transform(X_test)))
```

# Move onto Support Vector Regression

```python
home_sup = pd.read_csv('house_price_regression_dataset.csv')
X_supp = home_sup.iloc[:, :-1].values
y_supp = home_sup.iloc[:, -1].values.reshape(-1,1)
```

Feature scaling

```python
from sklearn.preprocessing import StandardScaler
sc_X_supp = StandardScaler()
sc_y_supp = StandardScaler()
X_supp = sc_X_supp.fit_transform(X_supp)
y_supp = sc_y_supp.fit_transform(y_supp)
```

Split the data

```python
X_supp_train, X_supp_test, y_supp_train, y_sup_test = \
    train_test_split(X_supp, y_supp, test_size=0.2, random_state=0)
```

Train the SVR model

```python
from sklearn.svm import SVR
regressor = SVR(kernel='rbf')
regressor.fit(X_supp_train, y_supp_train)
```

Predicting results

```python
sc_y_supp.inverse_transform(regressor.predict(X_supp).reshape(-1,1))
```
